# ahora we merge and work on the united front

```python
df={}
for i in rng:
    df[i] = pd.merge(pp[i], pc, on='Postcode')
```

```python
df[2019].head()
```

```python
boroughs = ['Camden',
'Greenwich',
'Hackney',
'Hammersmith and Fulham',
'Islington',
'City of London',
'Kensington and Chelsea',
'Lambeth',
'Lewisham',
'Southwark',
'Tower Hamlets',
'Wandsworth',
'Westminster',
'Barking and Dagenham',
'Barnet',
'Bexley',
'Brent',
'Bromley',
'Croydon',
'Ealing',
'Enfield',
'Haringey',
'Harrow',
'Havering',
'Hillingdon',
'Hounslow',
'Kingston upon Thames',
'Merton',
'Newham',
'Redbridge',
'Richmond upon Thames',
'Sutton',
'Waltham Forest']
```

```python
for i in rng:
    intq = df[i].Price.quantile(0.95)-df[i].Price.quantile(0.25)
    df[i] = df[i][df[i].District.isin(boroughs)]
    df[i] = df[i][df[i].Price >= (df[i].Price.quantile(0.25) - intq)]
    df[i] = df[i][df[i].Price <= (df[i].Price.quantile(0.75) + intq)]             
```

```python
df[1995]
```

```python
dfd = {}
for i in rng:
    dfd[i] = df[i].groupby('District', as_index=False).mean()
```

# ahora trabajamos con el GEOJSON de UK parlimentary wards

```python
gdf2 = geopandas.read_file('district.json')
```

```python
for i in rng:
    fig = px.choropleth(dfd[i], geojson=gdf2, locations='District', color='Price',
                        featureidkey="properties.LAD19NM",
                               color_continuous_scale="Blues")
    fig.update_geos(fitbounds="locations", visible=True)
    fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
    fig.write_html(r"C:\Users\Akiro\pp2-{}.html".format(i))
```

# ahora estudiamos los districts uno a uno

```python
dic = {}
for b in boroughs:
    temp=[]
    for i in rng:
        temp.append(dfd[i].where(dfd[i].District == b).Price.dropna().reset_index(drop=True)[0])
        dic[b] = temp
dic['Year'] = range(1995,2020)
```

```python
df2 = pd.DataFrame.from_dict(dic)
```

```python
df2.head()
```

```python
fig, axs = plt.subplots(7,5, figsize=(20, 30))


axs = axs.ravel()
x = range(1995,2020)
    
for n in range(33):
        y = df2[df2.columns[n]]
        axs[n].scatter(x, y)
        z = np.polyfit(x, y, 1)
        p = np.poly1d(z)
        axs[n].set_title(df2.columns[n])
        axs[n].plot(x,p(x),"r--")
plt.show()
```
